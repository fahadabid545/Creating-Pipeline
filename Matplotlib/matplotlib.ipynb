### This document presents data visualizations that compare feedback from multiple users utilizing Python's Matplotlib module.

### PLOT GRAPH

```python
import matplotlib.pyplot as plt    # Importing matplotlib module as plt

x=[1,3,5,7]
y=[2,6,8,9]

plt.plot(x,y, 
        color="black",             # Changing the default color of line into black 
        linestyle=":",             # Changing line style into dotted line 
        linewidth="2")             # Changing line width

plt.title("Example Title")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
```

### BAR PLOT

```python
import matplotlib.pyplot as plt
import numpy as np

x= ["PYTHON", "R", "PHP", "JAVA", "HTML"]
y= [85,65,76,89,87]
z= [60,54,61,76,79]

width=0.3                       # Defining width of the bar in width variable
i=np.arange(len(x))             # Making an array 'i' of length list 'x'
ii=[j+width for j in i]         # Making another list using list comprehension method by adding width of bar in 'i' 

plt.xlabel("Languages",
        fontsize=12, 
        fontname='Times New Roman')
plt.ylabel("Popularity/useage", 
        fontsize=12, 
        fontname='Times New Roman')
plt.title("Language Trend", 
        fontsize=15, 
        fontweight='bold', 
        fontname="Times New Roman",
        y=1.05)                # Lifting up title 

plt.bar(i,y,                  # using width variable for bae width 
        width, 
        color="pink",         # Bar color
        label="Popularity")

plt.bar(ii,z,
        width,              # using width variable for bae width 
        color="maroon",     # Bar color 
        label="Useage")

plt.xticks(i+width/2,       # adding the width of bar in array'i'
        x,                  # using the list 'x' instead of array 'i' for user readability
        rotation=45)        # Rotating the values on x-axis
plt.legend()
plt.show()
```

### SCATTER PLOT

```python
import matplotlib.pyplot as plt

x= x= ["PYTHON", "R", "PHP", "JAVA", "HTML"]
y= [85,65,76,89,87]
z= [60,54,61,76,79]

sizes1=[100,200,400,600,300]
sizes2=[300,600,400,200,100]
color1=[10,25,40, 90, 100]
color2=["r", "m", "g", "y", "y"]

plt.xlabel("Languages",fontsize=12, fontname='Times New Roman')
plt.ylabel("Popularity/useage", fontsize=12, fontname='Times New Roman')
plt.title("Language Trend", fontsize=15, fontweight='bold', fontname="Times New Roman")
    
scatter=plt.scatter(x,y, 
                    c=color1,            # Color grading in between 1-100
                    cmap="viridis",      # Choosing color map
                    s=sizes1)          
plt.scatter(x,z,
            c=color2 ,                   # Assigning colors directly
            s=sizes2)

color_bar=plt.colorbar(scatter)          # To see the color grading in color bar
color_bar.set_label("Color Scale")       # Labeling the color bar

plt.show()
```

### PIE CHART

```python
import matplotlib.pyplot as plt

x= ["PYTHON", "R", "PHP", "JAVA", "HTML"]
y= [60,54,61,76,79]

colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen', 'lavender']    # Colors respective to the labels "x"
explode_space=[0.2,0.0,0.0,0.0,0.0]                        # Exploding only python's wedge

plt.pie(y, explode=explode_space                           # Explode for better visuals
        , labels=x
        , shadow=True                                      # Add shadows to the diagram
        , colors=colors
        , autopct="%d%%"                                   # Will show the percentage of each language out of 100 as an integer
        , radius=1.2
        , labeldistance=1.15                               # Extended label distance for better readability
        , textprops={"fontname":"Times New Roman"}         # Adding Text properties  
        ,wedgeprops={"linewidth":5, "edgecolor":"black"})  # Adding wedge properties like increasing line width and edge color 
plt.title("Golbal Useage of Programming Languages", 
        fontname="Times New Roman", 
        y=1.1)
plt.legend(loc=5)                                          # Loc -> Selecting the location of the legend
plt.show()
```

### DONUT/RING PIE CHART

```python
import matplotlib.pyplot as plt

x= ["PYTHON", "R", "PHP", "JAVA", "HTML"]
y= [60,54,61,76,79]
explode_space=[0.2,0.0,0.0,0.0,0.00]        # Assigning explode values to wedge 

plt.pie(y, 
        explode=explode_space,              # Explode for better visuals
        labels=x, 
        radius=1.2)
plt.title("Golbal Useage of Programming Languages", 
        fontname="Times New Roman", 
        y=1.1)                              # To elevate the title

# Pie Chart within a Pie chart
plt.pie([1], 
        colors="w", 
        radius=0.8)

plt.show()
```

### ANOTHER WAY TO MAKE DONUT/RING PIE CHART

```python
import matplotlib.pyplot as plt

x= ["PYTHON", "R", "PHP", "JAVA", "HTML"]
y= [60,54,61,76,79]

plt.pie(y, 
        labels=x,              # Passing list 'x' as labels
        radius=1.2)
plt.title("Golbal Useage of Programming Languages",
        fontname="Times New Roman", 
        y=1.1)                 # To elevate the title

circle=plt.Circle(xy=(0,0),    # To create a circle object at axis(0,0)
                radius=0.8,    # Circle radius
                facecolor="w") # Circle Color

plt.gca().add_artist(circle)   # gca -> Stands for Get Current Axis (of Pie Chart) to add circle object in it.
plt.show()
```

### STEM PLOT

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5,6] 
y=[2,3,4,4,3,5]

plt.stem(x,y, 
        basefmt="y",                   # Assigning yellow color to base line 
        markerfmt="mD",                # Assigning marker shape D -> Diamond and color m -> Maroon
        linefmt="-.",                  # Changing default line style into dash and dot line
        orientation="vertical")        # Assigning orientation vertical which is also by default 
plt.show()
```

### BOX WHISKER PLOT

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5,6,9] 
y=[2,3,4,4,3,5,7]
z=[x,y]

plt.boxplot(z,                                 # Plotting 2 box whisker plots at 2 different locations
        tick_labels=["Label_1", "Label_2"],
        showmeans=True,                        # Mean value shown by green value 
        sym="rD",                              # Shows outlier in red color and diamond shape
        showfliers=True,                       # show outliers which is by default
        boxprops=dict(color="y"),              # changing box color
        capprops=dict(color="m"),              # Changing cap color
        capwidths=0.2,                         # Extending cap width 
        widths=0.2,                            # Extending box width
        whiskerprops=dict(color="b"))          # Changing whisker color
plt.show()
```

### STACK AREA PLOT

```python
import matplotlib.pyplot as plt

x=[1,2,3,6,7,7,9]                     # X-axis values           
area1=[2,4,4,5,7,9,7]                 # Y-axis values
area2=[1,7,5,3,4,3,2]
area3=[0,2,3,5,6,6,5]

# Area= x*y, therefore we used x*area1, x*area2 and x*area3 for 3 different areas
plt.stackplot(x,area1,area2,area3,
            labels=["area1", "area2", "area3"],
            colors=["m","g","y"])

plt.title("PYTHON")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```

### STEP PLOT

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5,6,9] 
y=[2,3,4,4,3,5,7]
plt.step(x,y,
        color="r",                   # Changing line color to red 
        marker="s",                  # Changing marker shape to square
        mfc="y",                     # Changing marker color to yellow
        ms=7)                        # Changing marker size
plt.title("LANGUAGES FEEDBAK COMPARISION")
plt.xlabel("Python")
plt.ylabel("Java")

plt.show()
```

### FILL BETWEEN

```python
import matplotlib.pyplot as plt    # Importing matplotlib module as plt
import numpy as np                 # Importing numpy module

x=np.array([1,3,5,7])
y=np.array([2,6,8,9])

plt.plot(x,y, 
        color="black",             # Changing the default color of line into black 
        linestyle=":",             # Changing line style into dotted line 
        linewidth="2")             # Changing line width

plt.title("Example Title")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.fill_between(x,y, color="y", 
                where=(x>=1)&(x<=5))       # where condition is only applicable on arrays
plt.show()
```

### SUBPLOTS

```python
import matplotlib.pyplot as plt 

x=[1,2,3,4] 
y=[2,3,4,4]
z=["Python","Php","Java","Html"]
plt.subplot(2,2,1)        # Make a plot with 2 rows and 2 columns 
plt.plot(x,y)             # adding as subplot at index 1

plt.subplot(2,2,2)
plt.pie(y)               # adding as subplot at index 2

plt.subplot(2,2,3)
plt.pie([1])             # adding as subplot at index 3

plt.subplot(2,2,4)
plt.bar(z,y)             # adding as subplot at index 4
```
