##### As the data Shows we have 20 movies on each page and we have 502 pages in total, therefore we will fetch the specific key in the data which is 'result' here which shows the result on the present page, and then make a loop to get the result from each page, and then save it in data frame


```python
API_KEY = os.getenv("TMDB_API_KEY") # Ensure Your API Key
dataset = pd.DataFrame()
all_pages = []  

for i in range(1, 502): 
    url = f"https://api.themoviedb.org/3/movie/top_rated?language=en-US&page={i}&api_key={API_KEY}"
    headers = {"accept": "application/json"}
    
    response = requests.get(url, headers=headers)
    
    # Check if the request was successful
    if response.status_code == 200:
        data = response.json()  
        
        # Check if 'results' key exists in response
        if 'results' in data:
            page_result = pd.DataFrame(data['results'])[['id', 'title', 'overview', 'release_date', 'popularity', 'vote_average', 'vote_count']]
            all_pages.append(page_result)  
        else:
            print(f"❌ 'results' key missing on page {i}")
            print(data)  
            break  
    else:
        print(f"❌ API Request Failed on page {i} | Status Code: {response.status_code}")
        print(response.text)  
        break

# Combine all pages into a single DataFrame
if all_pages:
    dataset = pd.concat(all_pages, ignore_index=True)
    print("✅ Data collection complete!")
else:
    print("❌ No valid data retrieved.")

print(dataset.shape)  
dataset
```

```python
dataset.to_csv('TMDB Top Movies.csv')
```
